```python
grades = [85, 90, 78, 92, 88]

count = 0
total = 0
for grade in grades:
    count+=1
    total+=grade

average = total/count
max = sorted(grades, reverse=True)[0]
min = sorted(grades)[0]

excellent_count = 0
for grade in grades:
    if grade >= average:
        excellent_count+=1

print(f"平均点:{average}")
print(f"最高点:{max}")
print(f"最低点:{min}")
print(f"平均以上の学生数:{excellent_count}")
```
