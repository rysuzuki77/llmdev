```python
alphabet_list = list("abcdefghijklmnopqrstuvwxyz")
alphabet_counts = dict.fromkeys(alphabet_list, 0)
word_list = []
while True: 
    word = input("英単語を入力してください")
    if word == "":
        break
    print(f"英単語を入力してください:{word}")
    word_list.append(word)
    for alphabet in alphabet_counts:
        alphabet_counts[alphabet] += word.count(alphabet)

word_list.sort()
print(f"入力した英単語:{word_list}")

for alphabet, counts in alphabet_counts.items():
    if counts > 0:
        print(f"{alphabet}が{counts}個ありました")
```
