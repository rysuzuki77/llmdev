```python
class BankAccount:

    def __init__(self, amount:int):
        self.amount = amount

    def deposit(self, amount:int):
        if amount < 0:
            print("エラー: 正しい金額を入力してください")
        self.amount += amount
        return True

    def withdraw(self, amount:int):
        if self.amount < amount:
            print("エラー: 残高不足または無効な金額です。")
        self.amount -= amount
        return True

    def get_balance(self):
        return self.amount

amount = BankAccount(1000)
amount.deposit(500)
amount.withdraw(200)
print(amount.get_balance())
amount.withdraw(1500)
```
