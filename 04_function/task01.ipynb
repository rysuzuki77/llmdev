```python
def is_valid_username(name):
    if len(name) < 5 or len(name) > 15:
        print("エラー: ユーザー名は5文字以上15文字以内で入力してください")
        print("不正なユーザー名です")
        return False
    if " " in name:
        print("エラー: ユーザー名に空白を含めることはできません")
        print("不正なユーザー名です")
        return False
    if not name.isalnum():
        print("エラー: ユーザー名は英数字のみ使用できます")
        print("不正なユーザー名です")
        return False

    print("ユーザー名は有効です")
    return True

is_valid_username("JohnDoe123")
is_valid_username("JD")
is_valid_username("John@Doe")
is_valid_username("John Doe")
```
